```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df=df[['survived','pclass','age','sibsp','parch','fare','sex']]
```

```python
sns.pairplot(df)
```

```python
sns.rugplot(df['fare'])
```

```python
sns.barplot(x='sex',y='age',data=df)
```

```python
sns.countplot(x='sex',data=df)
```

```python
sns.boxplot(x='sex',y='age',data=df)
```

```python
sns.boxplot(x='sex',y='age',data=df,hue='survived')
```

```python
sns.violinplot(x="sex",y='age',data=df)
```

```python
sns.violinplot(x="sex",y='age',data=df,hue='survived')
```

```python
sns.stripplot(x='sex',y='age',data=df,jitter=True)
```

```python
sns.stripplot(x='sex',y='age',data=df,hue='survived',jitter=True)
```

```python
sns.swarmplot(x='sex',y='age',data=df)
```

```python
sns.swarmplot(x='sex',y='age',data=df,hue="survived")
```

```python
sns.swarmplot(x='Sex',y='Age',data=df,hue="Survived",legend=False)
```

```python
sns.violinplot(x='sex', y='age', data=df)
sns.swarmplot(x='sex', y='age', data=df, color='black')
```
