```python
sns.countplot(x='species',data=df)
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=df)
```

```python
sns.scatterplot(x='petal_length',y='petal_width',hue='species',data=df)
```

```python
sns.pairplot(df,hue='species')
```

```python
fig ,axes=plt.subplots(2,2,figsize=(10,10))
axes[0,0].set_title("Sepal Length")
axes[0,0].hist(df['sepal_length'])
axes[0,1].set_title('Sepal Width')
axes[0,1].hist(df['sepal_width'])
axes[1,0].set_title('Petal Length')
axes[1,0].hist(df['petal_length'])
axes[1,1].set_title('Petal width')
axes[1,1].hist(df['petal_width'])
```

```python
plot=sns.FacetGrid(df,hue='species')
plot.map(sns.distplot,"sepal_length").add_legend()

plot=sns.FacetGrid(df,hue='species')
plot.map(sns.distplot,"sepal_width").add_legend()

plot=sns.FacetGrid(df,hue='species')
plot.map(sns.distplot,"petal_length").add_legend()

plot=sns.FacetGrid(df,hue='species')
plot.map(sns.distplot,"petal_width").add_legend()
```

```python
df_corr=df.drop(['species'],axis=1).corr(method='pearson')
```

```python
sns.heatmap(df_corr,annot=True)
```

```python

plt.figure(figsize=(10,10))
	
plt.subplot(221)
sns.boxplot(x="species", y='sepal_length', data=df)

plt.subplot(222)
sns.boxplot(x="species", y='sepal_width', data=df)

plt.subplot(223)
sns.boxplot(x="species", y='petal_length', data=df)

plt.subplot(224)
sns.boxplot(x="species", y='petal_width', data=df)

plt.show()
```

```python
Q1=np.percentile(df['sepal_width'],25,interpolation='midpoint')
Q3=np.percentile(df['sepal_width'],75,interpolation='midpoint')
```

```python
IQR=Q3 - Q1
```

```python
IQR
```

```python
print("Old Shpe ---->",df.shape)
```

```python
high=np.where(df['sepal_width']> Q3 + 1.5*IQR)
```

```python
low=np.where(df['sepal_width']< Q1 -1.5*IQR)
```

```python
high
```

```python
low
```

```python
df.drop(high[0],inplace=True)
```

```python
df.drop(low[0],inplace=True)
```

```python
print("New Shape ------>",df.shape)
```

```python
sns.boxplot(x='sepal_width',data=df)
```
