```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
stop_words=stopwords.words('english')
from nltk.stem import WordNetLemmatizer
from nltk.stem import SnowballStemmer
from nltk import FreqDist
from nltk import pos_tag
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text ="Text Analytics is the process of analyzing unstructured text data to extract relevant information and insights It involves several preprocessing steps such as tokenization, POS tagging, stop words removal, stemming, and lemmatization. Once the text is preprocessed, we can create representations of documents using techniques like TF-IDF."
```

```python
text=text.lower()
```

```python
#Tokenization
#  Tokenization is the process of breaking down a text into words or sentences.
#  The nltk library provides the sent_tokenize and word_tokenize functions to tokenize text into sentences and words respectively.
#Word Tokenization
words=word_tokenize(text)
print(words)
#Sentence Tokenization
sentences=sent_tokenize(text)
print(sentences)
```

```python
stop_words
```

```python
#Stop Words Removal
Stopwords_Removal=[word for word in words if word not in stop_words]
print(Stopwords_Removal)
```

```python
Word_lemmatizer=WordNetLemmatizer()
lemmatized_words=[Word_lemmatizer.lemmatize(word) for word in Stopwords_Removal]
lemmatized_words
```

```python
Snowball_Stemmer=SnowballStemmer('english')
stemmed_words=[Snowball_Stemmer.stem(word)for word in Stopwords_Removal]
stemmed_words
```

```python
# Punctuation Removal
Words_without_punctuation=[word for word in lemmatized_words if word.isalnum()]
Words_without_punctuation
```

```python
#POS Tagging
pos_tag=nltk.pos_tag(Words_without_punctuation)
pos_tag
```

```python
# Frequency Distribution 
freq=FreqDist(Words_without_punctuation)
print(freq.most_common(10))
freq.plot(10)
```

```python
#Plotting the WordCloud
wordcloud=WordCloud().generate(text)
plt.figure(figsize=(12,12))
plt.imshow(wordcloud)
plt.axis("off")
plt.show
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
d0 = 'Optimus prime'
d1 = 'Megatron rules out the optimus prime'
d2 = 'descepticons arrives for optimus prime'
string =[d0,d1,d2]
string
```

```python
Tfidf=TfidfVectorizer()
Result=Tfidf.fit_transform(string)
```

```python
print(Result)
```

```python
#get idf values
print('\n idf values :')
for ele1,ele2 in zip(Tfidf.get_feature_names_out(),Tfidf.idf_):
    print(ele1,":",ele2)
```
