# Write a Python program to display some basic statistical details like percentile, mean, standard deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris-versicolor’ of iris.csv dataset. Provide the codes with outputs and explain everything that you do in this step.

```python
df_iris=pd.read_csv('iris.csv')
```

```python
df_iris
```

```python
print("Iris-Setosa")
setosa = df_iris['species']=='Iris-setosa'
df_iris[setosa].describe()
```

```python
print("Iris-versicolor")
setosa = df_iris['species']=='Iris-versicolor'
df_iris[setosa].describe()
```

```python
print("Iris-virginica")
setosa = df_iris['species']=='Iris-virginica'
df_iris[setosa].describe()
```

```python
print("Minimum in the feature Sepal Length ---->",df_iris['sepal_length'].min())
print("Minimum in the feature Sepal Width ---->",df_iris['sepal_width'].min())
print("Minimum in the feature Petal Length ---->",df_iris['petal_length'].min())
print("Minimum in the feature Petal Width ---->",df_iris['petal_width'].min())
```

```python
print("Maximum in the feature Sepal Length ---->",df_iris['sepal_length'].max())
print("Maximum in the feature Sepal Width ---->",df_iris['sepal_width'].max())
print("Maximum in the feature Petal Length ---->",df_iris['petal_length'].max())
print("Maximum in the feature Petal Width ---->",df_iris['petal_width'].max())
```

```python
print("Standard Deviation of Sepal Length ---->",df_iris['sepal_length'].std())
print("Standard Deviation of Sepal Width ---->",df_iris['sepal_width'].std())
print("Standard Deviation of  Petal Length ---->",df_iris['petal_length'].std())
print("Standard Deviation of  Petal Width ---->",df_iris['petal_width'].std())
```

```python
print("Q1-Percentile of feature Sepal Length ---->",np.percentile(df_iris['sepal_length'],25))
print("Q1-Percentile of  feature Sepal Width ---->",np.percentile(df_iris['sepal_width'],25))
print("Q1-Percentile of feature Petal Length ---->",np.percentile(df_iris['petal_length'],25))
print("Q1-Percentile of  feature Petal Width ---->",np.percentile(df_iris['petal_width'],25))
```

```python
print("Q2-Percentile of feature Sepal Length ---->",np.percentile(df_iris['sepal_length'],50))
print("Q2-Percentile of  feature Sepal Width ---->",np.percentile(df_iris['sepal_width'],50))
print("Q2-Percentile of feature Petal Length ---->",np.percentile(df_iris['petal_length'],50))
print("Q2-Percentile of  feature Petal Width ---->",np.percentile(df_iris['petal_width'],50))
```
